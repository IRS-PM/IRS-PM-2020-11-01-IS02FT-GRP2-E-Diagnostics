### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing the Dataset

```python
data = pd.read_csv('/content/gdrive/My Drive/RS/symptoms_diseases.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data = data.drop(columns=['Unnamed: 74'], axis = 1)
```

```python
data.head()
```

```python
X = data.drop(["Disease"],axis = 1)
y = data['Disease']
```

```python
X.shape
```

```python
data.info()
```

```python
# Splitting into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0, test_size= 0.20)
```

```python
print('Training Set: ', len(X_train))
print('Test Set: ', len(X_test))
print('Training Labels: ', len(y_train))
print('Test Labels: ', len(y_test))
```

### Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
# Confusion Matrix
y_pred = nb.predict(X_test)
print(confusion_matrix(y_test, y_pred))  
print(classification_report(y_test, y_pred)) 
```

```python
print('Accuracy Score: '+ str(accuracy_score(y_test, y_pred)*100)+ ' %')
```

### Prediction

```python
symptoms_categoty = {}

for index, symptom in enumerate(X):
    symptoms_categoty[symptom] = index
```

```python
symptoms_categoty
```

```python
input_vector = np.zeros(len(symptoms_categoty))
```

```python
nb.predict([input_vector])[0]
```
